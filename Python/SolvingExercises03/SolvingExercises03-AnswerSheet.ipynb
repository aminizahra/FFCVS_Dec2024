Class and Object
Implement following architecture using OOP!
use __str__ to print output same as: Breed of the dog is {Breed}, It is {Size} and {Color} , the dog is {Age} years old.
</div>

```python
class Dog:
    def __init__(self, name, breed, size, age, color):
        self.name = name
        self.breed = breed
        self.size = size
        self.age = age
        self.color = color
    
    def eat(self):
        return f"{self.name} is eating."
    
    
    def sleep(self):
        return f"{self.name} is sleaping."
    
    def sit(self):
        return f"{self.name} is sitting."
    
    def run(self):
        return f"{self.name} is running."
    
    def __str__(self):
        return f"Breed of the dog is {self.breed}, It is {self.size} and {self.color} , the dog is {self.age} years old."
```

```python
#Ex
B = Dog(name='Bella', breed='X', size='Medume', age=3, color='brown')
```

```python
str(B)
```
